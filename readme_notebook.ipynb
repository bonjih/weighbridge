## Usecase
To detect truck number and letters as they pass in and out of the weighbridge.

## Constrants
Limited video to construct a dataset, dataset will take a long to compile.

## Solution
Use license free pretrained dataset for segmentaion and OCR.<br>
For segmentation, the pretrained models and weights come from Enet (sementic), which trained on a Cityscapes dataset. For OCR, a combination of CRAFT (craft_mlt_25k.pth) with pardeq.pt (pretrained STR datasets) https://github.com/baudm/parseq/tree/main/strhub.


MIT - https://en.wikipedia.org/wiki/MIT_License <br>
Appachie 2.0 - https://en.wikipedia.org/wiki/Apache_License

## Implementaion

- The trucks are detected (by the way of moving) when passing through ROI's (at both gates).
- A wait period is added for the truck to stop, pass through the gate. This is to reduce the noise, focus on the haul.
- Multipule segnetaions are done every 30 frames, again to reduce the noise. The frame with the least amount of 'black', OCR is performed
- Processing is only done when a truck is detected withing the ROI's
- Results for each analysis are saved in audit_data_csv.

### Assumptions

1. The text to be read is on the side or back of the truck.
2. Capturing every 30 frames appears to get the best result without reading and processing every frame. 
3. ROI1 haulage takes longer to come into view over ROI2.
4. Added (ROI2 has reduced gate weight period)
